```python
#文件名:
daily = ['dailydata1207_0290000314708.txt',
'dailydata1207_0290000317838.txt',
'dailydata1207_0290000328083.txt',
'dailydata1208_0290000328083.txt']

meterid = ['meterid1207_0290000314708.txt',
'meterid1207_0290000317838.txt',
'meterid1207_0290000328083.txt',
'meterid1208_0290000328083.txt']

monthlydata = ['monthlydata1207_0290000314708.txt',
'monthlydata1207_0290000317838.txt',
'monthlydata1207_0290000328083.txt',
'monthlydata1208_0290000328083.txt']

simumeterid = ['simumeterid1207_0290000314708.txt',
'simumeterid1207_0290000317838.txt',
'simumeterid1207_0290000328083.txt',
'simumeterid1208_0290000328083.txt']
```

```python
#打开CSV文件：
csvFile = open("../data/wholeDaily.csv",'w',newline = '')
csvWriter = csv.writer(csvFile)
```

```python
#先获取所有表ID：
meterIdList = []
for i in range(4):
    idFile = open('../data/Archive/'+meterid[i])
    lines = idFile.readlines()
    idNumber = len(lines[0].split())
    #1个表号位+169天+1个标记位 = 171列
    dailyTable = np.zeros((idNumber,171),"float64")
    print(dailyTable.shape)
    for line in lines:
        ids = line.split()
        for j in range(len(ids)):
            dailyTable[j,0] = ids[j]  #写入id
    
    dailyFile = open('../data/Archive/'+daily[i])
    lines = dailyFile.readlines()
    for j in range(len(lines)):  
        line = lines[j]
        all_readout = line.split("\t")[2].split()
        for k in range(idNumber):
            dailyTable[k,j+1] = all_readout[k]
        
    labelFile = open('../data/Archive/'+simumeterid[i])
    lines = labelFile.readlines()
    for line in lines:
        ids = line.split()
        for id_ in ids:
            for i in range(dailyTable.shape[0]):
                if float(id_) == dailyTable[i,0]:
                    dailyTable[i,171] = 1
        
        
    csvWriter.writerows(dailyTable) #最后写入csv中
```
